```python
from langchain_ollama import ChatOllama

# 객체 생성
llm = ChatOllama(
    temperature=0,  # 창의성 (0.0 ~ 2.0)
    model="gemma3:270m",  # 모델명
)

# 질의내용
question = "대한민국의 수도는 어디?"

# 질의
# print(f"[답변]: {llm.invoke(question)}")
response = llm.invoke(question)
```

```python
# LogProb 활성화
# 주어진 텍스트에 대한 모델의 토큰 확률의 로그 값 을 의미합니다. 토큰이란 문장을 구성하는 개별 단어나 문자 등의 요소를 의미하고, 확률은 모델이 그 토큰을 예측할 확률을 의미합니다.

# 객체 생성
llm_with_logprob = ChatOllama(
    temperature=0.1,  # 창의성 (0.0 ~ 2.0)
    max_tokens=2048,  # 최대 토큰수
    model="gemma3:270m",  # 모델명
).bind(logprobs=True)
```

```python
# 질의내용
question = "대한민국의 수도는 어디?"

# 질의
response = llm_with_logprob.invoke(question)
```

```python
# 결과 출력
response.response_metadata
```

```python
# 스트리밍 출력, 스트림 방식으로 질의, 질의에 대한 답변을 실시간으로 받을 때 유용
# answer 에 스트리밍 답변의 결과를 받습니다.
answer = llm.stream("대한민국의 아름다운 관광지 3곳과 주소를 노출")

# 스트리밍 방식으로 각 토큰을 출력합니다. (실시간 출력)
for token in answer:
    print(token.content, end="", flush=True)
```
